### 3.2 Creating the Board

```python
 # Implement code for 3.2 here
def create_board(size):
    """
    This function create a 2D list according to size
    
    Parameters
    ----------
    size, data type: integer
    
    Returns
    -------
    board, data type: 2D list
    
    """
    # Converting string input into integer
    board = []
    for i in range(size):
        row = [0]*size
        board.append(row)
        
    return board
```

### 3.3 Is the target position occupied?

```python
 # Implement code for 3.3 here
def is_occupied(board, x, y):
    """
    This function determine whether the input position is occupied
    
    Parameters
    ----------
    board, data type: 2D list
    x, data type: integer 
    y, data type: integer
    
    Returns
    -------
    True/False, Boolean
    """
    
    # x:rows,y:column, 0 indicate position is'nt occupied by stones
    return board[x][y] != 0
```

### 3.4 Placing a Stone at a Specific Intersection

```python
 # Implement code for 3.4 here
def place_on_board(board, stone, position):
    """
    This function place stone on the board based on position.
    For stones, 1 indicate "●",2 indicate "○".
    
    Parameters
    ----------
    board, data type: 2D list
    stone, data type: integer
    position, data type: tuple of two strings,e.g. ("0","B")
    
    Returns
    -------
    True/False, Boolean
    """
    # Convert tuple of strings to tuple of integers
    position = (int(position[0]),ord(position[1])-65)
    
    # Check if the position is valid and within the board size
    if position[0]<0 or position[0]>(len(board)-1) or position[1]<0 or position[1]>(len(board)-1):
        print("Position entered have exceeded the existing board. Please input again!")
        return False

    elif is_occupied(board, position[0],position[1]):
        print("Position has been occupied. Please input again!")
        return False
    
    elif 0 <= position[0] <= len(board) or 0 <= position[1] <= len(board):
        board[position[0]][position[1]]=stone
        return True
    
    else:
        print("Invalid input. Please input again!")
        return False
```

### 3.5 Printing the Board

```python
 # Implement code for 3.5 here
def print_board(board):
    """
    This function prints the board with stones.
    
    Parameters
    ----------
    board, data type: 2D list
    
    """
    size = len(board)
    #Print empty board
    #Capital Alphabets start at 65 in ASCII
#     for i in range(65, 65 + size):
#         #end function does not create new line
#         print(chr(i), end="  ")
#     print()
    
    i = 65
    while i < (65 + size):
    #end function does not create new line
        print(chr(i), end="  ")
        i+=1
    print()
    
    # i represents the row index 
    for i in range(size):
        # j represents the column index 
        for j in range(size):
            # 1 indicates stone for player 1
            if board[i][j]==1:
                print("●",end="")
            # 2 and 3 indicate stone for player 2 and computer player
            elif board[i][j]==2 or board[i][j]==3:
                print("○",end="")
            # 0 indicate position not occupied
            elif board[i][j]==0:
                print(" ",end="")
            print('',end="")
            
            # Stop printing when j reach board size
            if j is not (size-1):
                print('--',end="")
            else:
                print(" ",i,end="")
        print()
        
        #Stop printing when i reach board size
        if i is not (size-1):
            print('  '.join(['|']*(size)))
    
```

### 3.6 Check Available Moves

```python
 # Implement code for 3.6 here
def check_available_moves(board):
    """
    This function checks the available moves on the board
    
    Parameters
    ----------
    board, data type: 2D list
    
    Returns
    -------
    available_moves, data Type: a list of tuples, two strings inside tuples,e.g.[('0', 'A')]
    """
    # Create empty list for available moves
    available_moves = []
    size = len(board)
    
    # Loops there rows and columns of the 2D list to indicate unoccupied boards
    for i in range(size):
        for j in range(size):
            if board[i][j] == 0:
                position=(str(i),chr(65+j))
                available_moves.append(position)
            else:
                continue
    return available_moves
    
```

### 3.7 Check for the Winner

```python
 # Implement code for 3.7 here
def check_for_winner(board):
    """
    This function shows identify winner of the game.
    If continuous 5 stones in same color achieved, return the corresponding color
    Continuous 5 stones include horizontal, vertical and diagonal
    If board is full but no continuous 5 stones, return draw
    If no one won the game, return None
    
    Parameters
    ----------
    board, data type: 2D list
    
    Returns
    -------
    return "Black" when player 1 wins
    return "White" when player 2 or computer player wins
    return "Draw" when board is full and no winner
    return None if none requirements met
    """
    
    size = len(board)
    
    # Number of stones to win
    num_win = 5
    for player in (1,2,3):
        #Check for horizontal and vertical pattern
        for i in range(size):
            for j in range(size-(num_win-1)):
                if all(board[i][j+k]==player for k in range(num_win)) \
                    or all(board[j+k][i]==player for k in range(num_win)):
                    if player == 2 or player == 3:
                        return "White"
                    elif player == 1:
                        return "Black"

        #Check for diagonal pattern
        for i in range(size-(num_win-1)):
            for j in range(size-(num_win-1)):
                if all(board[i+k][j+k]==player for k in range(num_win)) \
                    or all(board[i+k][j+num_win-1-k]==player for k in range((num_win))):
                    if player == 2 or player == 3:
                        return "White"
                    elif player == 1:
                        return "Black"
                    
    # Check if board is full
    if all(board[i][j] != 0 for i in range(size) for j in range(size)):
        return "Draw"
    
    # None of the requirements met return None
    return None
```

### 3.8 Random Computer Player

```python
 # Implement code for 3.8 here
def random_computer_player(board, player_move):
    """
    This function shows generates a random move based on player's last move.
    The random move will be made on the surrounding available positions of player's 
    last move on the board.
    
    Step to optain the random move:
    1. Determine surrounding positions
    2. Available positions on the board
    3. Find intersection between these positions
    
    Parameters
    ----------
    board, data type: 2D list
    player_move, data type: tuples with two strings, e.g. ('0', 'A')
    
    Returns
    -------
    available_moves, data Type: a list of tuples, two strings inside tuples
    """
    player_move = (int(player_move[0]),ord(player_move[1])-65)
    
    # Define all surrounding positions
    surrounding_position = [(player_move[0],player_move[1]-1),   #left
                            (player_move[0],player_move[1]+1),   #right
                            (player_move[0]+1,player_move[1]),   #bottom
                            (player_move[0]-1,player_move[1]),   #top
                            (player_move[0]+1,player_move[1]+1), #bottom right
                            (player_move[0]+1,player_move[1]-1), #bottom left
                            (player_move[0]-1,player_move[1]+1), #top right
                            (player_move[0]-1,player_move[1]-1)  #top left
                            ]
    
    # Remove position outside range of board size
    # Initialize i 
    i = 0
    #Check each tuple in surrounding_position list, if -ve value spotted, remove the tuple
    while i < len(surrounding_position):
        if surrounding_position[i][0]< 0 or surrounding_position[i][1]< 0:
            surrounding_position.remove(surrounding_position[i])
        else:
            surrounding_position[i] = (str(surrounding_position[i][0]),chr(surrounding_position[i][1]+65))
            i+=1

    # Determine surrounding positions that is available
    # Initialize new list for surrounding available positions
    surrounding_available = []
    available_moves = check_available_moves(board)
    for i in surrounding_position:
        if i in available_moves:
            surrounding_available.append(i)

    # Computer player will choose any from the surrounding_position
    # Create a random index
    if surrounding_available:
        random_index = random.randint(0,len(surrounding_available)-1)
        random_position = surrounding_available[random_index]
    else:
        random_index = random.randint(0,len(available_moves)-1)
        random_position = available_moves[random_index]

    return random_position
```
